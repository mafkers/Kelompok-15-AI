# EDA Dataset UGD & Keluhan NLP
Eksplorasi awal untuk melihat sebaran data penjadwalan pasien dan data latih teks keluhan.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data Penjadwalan
df_ugd = pd.read_csv('dataset_ugd.csv')
print("=== Info Dataset UGD ===")
print(df_ugd.info())

plt.figure(figsize=(6,4))
sns.countplot(x='tingkat_kegawatan', data=df_ugd, palette='viridis')
plt.title('Distribusi Tingkat Kegawatan Pasien (Penjadwalan)')
plt.show()

# 2. Load Data NLP (Keluhan)
df_keluhan = pd.read_csv('dataset_keluhan.csv')
print("\n=== Info Dataset Keluhan ===")
print(df_keluhan.info())

plt.figure(figsize=(6,4))
sns.countplot(x='kegawatan', data=df_keluhan, palette='magma')
plt.title('Distribusi Label Kegawatan pada Teks Keluhan')
plt.show()
```
